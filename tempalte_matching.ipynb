```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/mnunew.jpg")
temp_image = cv2.imread("/content/Screenshot 2021-11-16 113616.png")
```

```python
(tH,tW) = temp_image.shape[:2] #for x axis and y axis
```

```python
(tH,tW)
```

```python
print("test Image")
cv2_imshow(image)
print("Template Image")
cv2_imshow(temp_image)
```

```python
# convert both the image and template to grayscale
imageGray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
templateGray = cv2.cvtColor(temp_image, cv2.COLOR_BGR2GRAY)
# perform template matching
print("performing template matching...")
result = cv2.matchTemplate(imageGray, templateGray,	cv2.TM_SQDIFF_NORMED)
```

```python
threshold=0.25

(yCoords, xCoords) = np.where(result <= threshold)
clone = image.copy()
print("matched locationS")
# loop over our starting (x, y)-coordinates
for (x, y) in zip(xCoords, yCoords):
	# draw the bounding box on the image
	cv2.rectangle(clone, (x, y), (x + tW, y + tH), (255, 255, 0))
cv2_imshow(clone)
```
